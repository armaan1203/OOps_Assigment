```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
df
```

distribution of age among heart failure patients

```python
sns.scatterplot(x=df["age"], y=df["platelets"], alpha=0.7, color="blue")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Age Distribution of Heart Failure Patients", fontsize=14)
plt.show()
```

The death rate vary as the it is highest from 55 to 60 age and the death rate  goes decressing after the age of 60.

```python
male_count = df[df['sex'] == 1].shape[0]
female_count = df[df['sex'] == 0].shape[0]
```

```python
plt.pie([male_count,female_count],labels=['male','female'],colors = ['#1f77b4', '#ff7f0e'],autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x=df["age"], weights=df["platelets"], bins=20, kde=True, color="blue")
plt.title('Age Distribution of platelate')
plt.xlabel('PLatelets count')
plt.ylabel('Age')
plt.show()
```

```python
sns.scatterplot(x=df["serum_creatinine"], y=df["serum_sodium"], alpha=0.7, color="blue")
plt.title('Relation between sodium level and creatinine')
plt.xlabel('Creatine')
plt.ylabel('Sodium level')
plt.show()
```

Yes, there is a relation between Sodium level and creatinine as the creatinine level is low then the sodium level is mostly high.

```python
gender_hbp_counts = df.groupby("sex")["high_blood_pressure"].value_counts(normalize=True).unstack()
gender_hbp_counts.index = ["Female", "Male"]
plt.figure(figsize=(8, 5))
gender_hbp_counts.plot(kind="bar", stacked=True, colormap="Blues", figsize=(8,6))
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Proportion", fontsize=12)
plt.title("Prevalence of High Blood Pressure by Gender", fontsize=14)
plt.legend(["No High BP", "High BP"], title="Condition")
plt.xticks(rotation=0)
plt.show()
```

As the analysis concluded that the females have the more high blood pressure as compared to male.

```python
gender_hbp_counts = df.groupby("smoking")["DEATH_EVENT"].value_counts(normalize=True).unstack()
gender_hbp_counts.index = ["smoking", "non-smoking"]
plt.figure(figsize=(8, 5))
gender_hbp_counts.plot(kind="bar", stacked=True, colormap="Reds", figsize=(8,6))
plt.xlabel("smoking habit", fontsize=12)
plt.ylabel("Proportion", fontsize=12)
plt.title("Impact of smoking on heart failure", fontsize=14)
plt.legend(["Survived", "Heart Failure"], title="outcome")
plt.xticks(rotation=0)
plt.show()
```

There is no relation between somking and heart failure because the death rate in smoker and non smoker is equal.

```python
plt.figure(figsize=(10, 6))
sns.histplot(x=df["age"], weights=df["DEATH_EVENT"], bins=20, kde=True, color="blue")
plt.title('Age Distribution of death')
plt.xlabel('death rate')
plt.ylabel('Age')
plt.show()
```

The death is highest are at the age of 60.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["diabetes"], y=df["ejection_fraction"], palette="coolwarm")
plt.xticks([0, 1], ["No Diabetes", "Diabetes"])
plt.xlabel("Diabetes Status", fontsize=12)
plt.ylabel("Ejection Fraction", fontsize=12)
plt.title("Ejection Fraction Comparison Between Patients With and Without Diabetes", fontsize=14)
plt.show()
```

Ejection fractiojn in more in non diabitic as compared to diabitics.

```python
plt.figure(figsize=(12, 5))
sns.histplot(data=df, x='serum_creatinine', hue='DEATH_EVENT', kde=True, bins=30, palette=['green', 'red'], alpha=0.6)
plt.title('Distribution of Serum Creatinine Levels')
plt.xlabel('Serum Creatinine (mg/dL)')
plt.ylabel('Frequency')
plt.show()
```
